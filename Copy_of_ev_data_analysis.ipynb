#18.	Model Year Trend:
•	Analyze the trend in the number of registered electric vehicles by model year. Provide a line chart to show any increase or decrease over the years.


```python
model_year_counts = df.groupby('Model Year')['VIN (1-10)'].count()

plt.figure(figsize=(10, 6))
plt.plot(model_year_counts.index, model_year_counts.values)

plt.xlabel("Model Year")
plt.ylabel("Number of Registered Electric Vehicles")
plt.title("Trend in Registered Electric Vehicles by Model Year")

plt.xticks(rotation=45)

plt.show()

percentage_change = []
for i in range(1, len(model_year_counts)):
    change = ((model_year_counts.values[i] - model_year_counts.values[i-1]) / model_year_counts.values[i-1]) * 100
    percentage_change.append(change)

print("Percentage Change in Registered Electric Vehicles by Model Year:")
for i, change in enumerate(percentage_change):
    print(f"From {model_year_counts.index[i]} to {model_year_counts.index[i+1]}: {change:.2f}%")
```

# 20.	Legislative District and MSRP:
•	How does the average base MSRP vary across different legislative districts?


```python
avg_msrp_by_district = df.groupby('Legislative District')['Base MSRP'].mean().round(2)

print("Average Base MSRP by Legislative District:")
print(avg_msrp_by_district.to_string())

sorted_avg_msrp = avg_msrp_by_district.sort_values(ascending=False)

print("\nTop 10 Districts with Highest Average Base MSRP:")
print(sorted_avg_msrp.head(10))
plt.figure(figsize=(10, 6))
avg_msrp_by_district.plot(kind='bar')

plt.xlabel("Legislative District")
plt.ylabel("Average Base MSRP")
plt.title("Average Base MSRP by Legislative District")

plt.xticks(rotation=45)
plt.show()
```
